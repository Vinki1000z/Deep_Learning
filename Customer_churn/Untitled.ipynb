```python
columns_to_replace = ['OnlineSecurity', 'OnlineBackup']
df[columns_to_replace].replace('No internet service','No',inplace=True)
```

```python
df.replace({'No internet service':'No','No phone service':'No'},inplace=True)
```

```python
for columns in df:
    print('{} {}'.format(columns,df[columns].unique()))
```

```python
# df.replace({'No':0,'Yes':1},inplace=True)
# wrong way beacue it changes the value of other in which we need the no
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for columns in df:
    print('{} {}'.format(columns,df[columns].unique()))
```

One Hot Encoding

```python
df=pd.get_dummies(data=df,columns=['InternetService','Contract','PaymentMethod'])
```

```python
df
```

```python
df.dtypes
```

```python
df['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df.dtypes
```

```python
X=df.drop(['Churn'],axis=1)
y=df.Churn  
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
y.shape
```

```python
y_train.shape
```

```python
y_train
```

```python
model=keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=200)
```

```python
model.evaluate(X_test, y_test)
```

```python
def predict(x):
    return 1 if x>=0.5 else 0
```

```python
yp=model.predict(X_test)
```

```python
yp
```

```python
y_predict=[]
for temp in yp:
    y_predict.append(predict(temp))
```
