increasing the model accuracy

```python
def ANN(x_train,x_test,y_train,y_test,loss):
    model=keras.Sequential([
    keras.layers.Dense(56, input_shape=(26,), activation='relu'),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    model.fit(x_train, y_train, epochs=200)
    y_pred_temp=model.predict(x_test)
    y_pred=np.round(y_pred_temp)
    print('evalued model ',model.evaluate(X_test, y_test))
    print("confusion report: \n",classification_report(y_test,y_pred))
    return y_pred
    
```

```python
y_pred=ANN(X_train,X_test,y_train,y_test,"binary_crossentropy")
print(y_pred)
```

method 3: SMOTE

```python
df4=df.copy()
```

```python
x=df4.drop(['Churn'],axis=1)
y=df4.Churn
```

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(x, y)

y_sm.value_counts()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)
```

```python
y_train.shape()
```

```python
y_pred=ANN(X_train,X_test,y_train,y_test,"binary_crossentropy")
```
