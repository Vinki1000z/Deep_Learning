```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.cifar10.load_data()
```

```python
x_train.shape
```

```python
x_train=x_train/255
x_test=x_test/255
```

making the 2-D and 1-D array of the y_train

```python
y_train
```

```python
y_train_2D=y_train
y_test_2D=y_test
```

```python
y_train_1D=y_train.reshape(len(y_train))
y_test_1D=y_test.reshape(len(y_test))
```

```python
print(y_train_1D.shape)
print(y_train_2D.shape)
```

function of finder

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

making the ANN

```python
ann_2D = keras.models.Sequential([
        keras.layers.Flatten(input_shape=(32,32,3)),
        keras.layers.Dense(3000, activation='relu'),
        keras.layers.Dense(1000, activation='relu'),
        keras.layers.Dense(10, activation='sigmoid')    
    ])

ann_2D.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

ann_2D.fit(x_train, y_train_2D, epochs=5)
```

```python
ann_1D = keras.models.Sequential([
        keras.layers.Flatten(input_shape=(32,32,3)),
        keras.layers.Dense(3000, activation='relu'),
        keras.layers.Dense(1000, activation='relu'),
        keras.layers.Dense(10, activation='sigmoid')    
    ])

ann_1D.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

ann_1D.fit(x_train, y_train_2D, epochs=5)
```

```python
ann_1D.evaluate(x_test,y_test_2D)
```

```python
y_temp_predict_1D=ann_1D.predict(x_test)
```

```python

y_predict_1D=[np.argmax(i) for i in y_temp_predict]
```

```python
finder_1D(x_test,y_test_1D,55)
print("Y_predict_2D",classes[y_predict_1D[55]])
```

CNN start

```python
cnn_2D=keras.models.Sequential([
    # feature extraction
   keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3)),
   keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    
#     classifier
     keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
    
    
])
cnn_2D.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
cnn_2D.fit(x_train, y_train_2D, epochs=10)
```

```python
cnn_2D.evaluate(x_test,y_test_2D)
```

```python
y_temp_predict_cnn_2D=cnn_2D.predict(x_test)
y_predict_cnn_2D=[np.argmax(i) for i in y_temp_predict_cnn_2D]
finder(x_test,y_test_2D,1)
print("Y_predict_2D",classes[y_predict_cnn_2D[1]])
```

```python
cnn_1D=keras.models.Sequential([
    # feature extraction
   keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3)),
   keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    
#     classifier
     keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
    
    
])
cnn_1D.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
cnn_1D.fit(x_train, y_train_1D, epochs=10)
```

```python
cnn_1D.evaluate(x_test,y_test_1D)
```
