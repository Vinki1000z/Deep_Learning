```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x_train_flatter=x_train.reshape(len(x_train),28*28)
```

```python
x_test_flatter=x_test.reshape(len(x_test),28*28)
```

```python
x_train_flatter[0]
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train_flatter)
```

```python
x_train_flattern=scaler.transform(x_train_flatter)
```

```python
scaler.fit(x_test_flatter)
x_test_flatter=scaler.transform(x_test_flatter)
```

```python
x_test_flatter[0]
```

```python
x_train_flattern[0]
```

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_flatter, y_train, epochs=5)
```

```python
model.evaluate(x_test_flatter,y_test)
```
